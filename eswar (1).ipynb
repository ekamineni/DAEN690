```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Users/eswar/Downloads/MODEL VS ACCURACY.xlsx")
```

```python
df
```

```python
N = df.shape[0]
python = df.iloc[0:,1]
momentum_ai = df.iloc[0:,2]
ind = np.arange(N)

plt.figure(figsize=(20,15))
width = 0.3   

plt.xticks(rotation = 45)

plt.bar(ind, python , width, label='PYTHON',color = 'gold')
plt.bar(ind + width, momentum_ai, width, label='MOMEMNTUM AI',color = 'green')

plt.xlabel('MODEL ID')
plt.ylabel('ACCURAY')
plt.title('COMPARISION BETWEEN ACCURACIES FOUND FROM PYTHON Vs MOMENTUM AI')
plt.xticks(ind + width / 2, df.iloc[:,0])
plt.legend(loc='best')
plt.show()
```
